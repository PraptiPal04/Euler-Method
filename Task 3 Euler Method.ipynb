```python
import numpy as np
import matplotlib.pyplot as plt

def Euler(f,x0,T,h,*args,**kwargs):
    x=np.zeros(int(T/h)+1)
    x[0]=x0
    for i in range(len(x)-1):
        x[i+1]=x[i]+h*f(x[i],*args,**kwargs)
    return x

def ode(x,d,u,al,b):
    return -d*x+u*np.tanh(al*x)+b

T=7
```

```python
#Part a 
h=0.001
t=np.arange(0.0,T+h,h)
xp=Euler(ode,-1,T,h,d=1,u=0.31,al=1.2,b=0)
plt.plot(t,xp)
plt.xlabel("t")
plt.ylabel("x")
plt.show()
```

```python
#Part b
h=0.001
t=np.arange(0.0,T+h,h)
xp=Euler(ode,-1,T,h,d=1,u=0.31,al=1.2,b=3)
plt.plot(t,xp)
plt.xlabel("t")
plt.ylabel("x")
plt.show()
```

```python
#Part c 
h=0.1
t=np.arange(0.0,T+h,h)
xp=Euler(ode,-1,T,h,d=1,u=0.31,al=1.2,b=3)
plt.plot(t,xp)
plt.xlabel("t")
plt.ylabel("x")
plt.show()
```

```python
#Part d 
h=0.001
t=np.arange(0.0,T+h,h)
xp=Euler(ode,-1,T,h,d=4,u=0.31,al=1.2,b=0)
plt.plot(t,xp)
plt.xlabel("t")
plt.ylabel("x")
plt.show()
```

```python
#All in same plot
t1=np.arange(0.0,T+0.001,0.001)
t2=np.arange(0.0,T+0.1,0.1)
x1=Euler(ode,-1,T,0.001,d=1,u=0.31,al=1.2,b=0)
x2=Euler(ode,-1,T,0.001,d=1,u=0.31,al=1.2,b=3)
x3=Euler(ode,-1,T,0.1,d=1,u=0.31,al=1.2,b=3)
x4=Euler(ode,-1,T,0.001,d=4,u=0.31,al=1.2,b=0)
plt.plot(t1,x1,label="Part a")
plt.plot(t1,x2,label="Part b")
plt.plot(t2,x3,label="Part c")
plt.plot(t1,x4,label="Part d")
plt.xlabel("t")
plt.ylabel("x")
plt.legend()
plt.show()
```
